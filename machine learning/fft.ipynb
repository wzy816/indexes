```
import numpy as np
from matplotlib.pyplot import plot, show

#创建一个余弦波信号
x = np.linspace(0, 2 * np.pi, 30) 
wave = np.cos(x)
plot(wave)
```

```
# 使用fft函数进行傅里叶变换。
transformed = np.fft.fft(wave)

# ifft函数近似地还原初始信号。
i_transformd = np.fft.ifft(transformed)

# 看差别
print(np.all(np.abs(i_transformd - wave) < 10 ** -9))

plot(i_transformd)
show()
```
