```python
import torch
import torch.nn.functional as F

def softmax_one(x, dim):
    x = x - x.max(dim=dim, keepdim=True).values
    exp_x = torch.exp(x)
    return exp_x / (1 + exp_x.sum(dim=dim, keepdim=True))

softmax = F.softmax
```
