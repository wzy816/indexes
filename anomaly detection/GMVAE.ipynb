```python
import tensorflow as tf
print(tf.__version__)
```

```python
# params

class AttrDict(dict):
    def __init__(self, *args, **kwargs):
        super(AttrDict, self).__init__(*args, **kwargs)
        self.__dict__ = self
        
params =  AttrDict({
    "gaussian_size": 64,
    "hard_gumbel" :0,
    "loss_type" : "bce",
    "dataset" :  None
}) 
```

```python
from GMVAE.tensorflow.networks.Networks import Networks

tf.get_default_graph()

data = tf.placeholder(tf.float32, shape=(64, 784)) # ===[batch_size, n_features]

network = Networks(params)

latent_spec = network.encoder(data, 10, True)
out_logits, y_mean, y_var, output = network.decoder(latent_spec['gaussian'], latent_spec['categorical'], 784, True)
latent_spec['output'] = out_logits
latent_spec['y_mean'] = y_mean
latent_spec['y_var'] = y_var

tf.get_collection("trainable_variables")
```
